## Task 3: Data Visualization (Trends)

Create a line plot showing **GDP Growth (%)** and **Inflation (%)** over the years.
This will help us see if high growth periods correlate with high inflation.

```python
plt.figure(figsize=(12, 6))

# Plot GDP Growth
plt.plot(df['Year'], df['GDP_Growth_Pct'], marker='o', label='GDP Growth (%)')

# Plot Inflation
plt.plot(df['Year'], df['Inflation_Pct'], marker='s', label='Inflation (%)', linestyle='--')

plt.title('Ethiopia: GDP Growth vs Inflation (2000-2024)')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

## Task 4: Relationship Analysis

We often hear about the relationship between the Exchange Rate and Inflation. 
Let's visualize the **Exchange Rate** over time and check its correlation with other variables.

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Year', y='Exchange_Rate_Birr_USD', color='green')
plt.title('USD to ETB Exchange Rate Trend')
plt.ylabel('Birr per USD')
plt.show()
```

Now, calculate the correlation matrix to see how the variables relate to each other.

```python
# Calculate correlation
correlation_matrix = df.drop('Year', axis=1).corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Economic Indicators')
plt.show()
```

## Task 5: Structural Change in Economy

Ethiopia has been aiming to industrialize. Let's see if the share of Agriculture in GDP has decreased over time.

1. Plot `Agri_Value_Added_Pct_GDP` over `Year`.
2. Calculate the average agricultural share for the first decade (2000-2010) vs the last decade (2014-2024).

```python
# Plot Agriculture Share
plt.figure(figsize=(10, 5))
plt.fill_between(df['Year'], df['Agri_Value_Added_Pct_GDP'], color="skyblue", alpha=0.4)
plt.plot(df['Year'], df['Agri_Value_Added_Pct_GDP'], color="Slateblue", alpha=0.6, linewidth=2)

plt.title('Share of Agriculture in GDP over Time')
plt.ylabel('Value Added (% of GDP)')
plt.xlabel('Year')
plt.show()
```

```python
# Calculate averages
avg_agri_2000_2010 = df[(df['Year'] >= 2000) & (df['Year'] <= 2010)]['Agri_Value_Added_Pct_GDP'].mean()
avg_agri_2014_2024 = df[(df['Year'] >= 2014) & (df['Year'] <= 2024)]['Agri_Value_Added_Pct_GDP'].mean()

print(f"Average Agri Share (2000-2010): {avg_agri_2000_2010:.2f}%")
print(f"Average Agri Share (2014-2024): {avg_agri_2014_2024:.2f}%")
```
